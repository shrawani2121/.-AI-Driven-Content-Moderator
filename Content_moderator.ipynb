```python
from transformers import pipeline

# Load pre-trained model for text classification
moderator_model = pipeline("sentiment-analysis")
```

```python
def moderate_content(text):
    # Analyze the sentiment of the input text
    result = moderator_model(text)[0]

    # Determine if the content is inappropriate based on the sentiment
    if result['label'] == 'NEGATIVE' and result['score'] > 0.8:
        return "Inappropriate content detected and flagged."
    else:
        return "Content is appropriate."
```

```python
# Simulate real-time monitoring of content
def monitor_content_stream():
    print("Content Moderation System is running...")
    print("Type 'exit' to stop the system.")

    while True:
        # Simulate receiving content from a stream (e.g., social media posts, comments)
        content = input("Enter content to moderate: ")

        if content.lower() == 'exit':
            print("Stopping Content Moderation System.")
            break

        # Moderate the content
        moderation_result = moderate_content(content)
        print(f"Moderation Result: {moderation_result}")

# Run the content monitoring simulation
monitor_content_stream()
```
